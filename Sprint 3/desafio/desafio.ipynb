#### Os 5 maiores aplicativos por números de instalação 

```python
df_google['Installs'] = df_google['Installs'].astype(str)

# Remover o valor 'Free' e o símbolo '+'
df_google['Installs'] = df_google['Installs'].str.replace('Free', '', regex=False)
df_google['Installs'] = df_google['Installs'].str.replace('+', '', regex=False)
df_google['Installs'] = df_google['Installs'].str.replace(',', '', regex=False)

# Substituir strings vazias por NaN
df_google['Installs'] = df_google['Installs'].replace('', np.nan)

# Converter a coluna 'Installs' de string para float
df_google['Installs'] = df_google['Installs'].astype(float)
df_google['Installs'] = pd.to_numeric(df_google['Installs'], errors='coerce')

# Ordenar os aplicativos pela coluna 'Installs' em ordem decrescente
top_5_apps = df_google.sort_values(by='Installs', ascending=False).head(5)

print(top_5_apps[['App', 'Installs']])
```

#### Representação gráfica em formato de barra dos 5 maiores aplicativos por número de instalação

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

print(top_5_apps[['App', 'Installs']])

plt.figure(figsize=(10, 6))

plt.bar(top_5_apps['App'], top_5_apps['Installs'], color=['blue', 'orange', 'green', 'red', 'purple'], width=0.4)

# Definir o rótulo do eixo Y
plt.ylabel('Número de Instalações')

plt.xlabel('Aplicativos')


plt.title('Top 5 Aplicativos por Número de Instalações')

plt.ylim(0, top_5_apps['Installs'].max() + 1000000)

plt.xticks(rotation=45)

ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x)))

plt.tight_layout()

plt.show()
```

#### Contar quantos aplicativos existem em cada categoria
